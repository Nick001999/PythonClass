```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
flights_data = pd.read_csv('/Users/nikolaykarabakin/Desktop/flight.csv')
```

```python
flights_data.shape
```

```python
flights_data.head()
```

```python
features = ["month", "origin", "arr_delay"]
data = flights_data[features]
```

```python
data = pd.get_dummies(data, columns = ["origin", "month"])
```

```python
data.head()
```

```python
data["delayed"] = (data["dep_delay"] > 0).astype(int)
data.drop("dep_delay", axis=1, inplace=True)
```

```python
data["delayed"] = (data["arr_delay"] > 0).astype(int)
data.drop("arr_delay", axis=1, inplace=True)
```

```python
data["delayed"] = (data["dep_delay"] > 0).astype(int)
data.drop("dep_delay", axis=1, inplace=True)
```

```python
data.head()
```

```python
X = data.drop("delayed", axis=1)
y = data["delayed"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=80)
```

```python
model = RandomForestClassifier(random_state=80)
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)

accuracy = accuracy_score(y_test, prediction)
```

```python
print(f"Accuracy of the model: {accuracy:.1%}")
```
